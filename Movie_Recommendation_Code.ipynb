```python
import pandas as pd
import numpy as np
import ast # Abstract syntax tree
```

```python
movies = pd.read_csv("movies.csv")
movies.head(1)
```

```python
credits = pd.read_csv("credits.csv")
credits.head(1)
```

```python
data = movies.merge(credits,on = "title")
data.head(1)
```

```python
data = data[["id","title","genres","keywords","overview","cast","crew"]]
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace = True)
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.iloc[0].genres # it is like a  string enclosed list of dictionary
```

```python
def convert(obj):
    lst = []
    for i in ast.literal_eval(obj):    # literal_eval() used to convert string enclosed list to the list
        lst.append(i["name"])
    return lst
```

```python
data["genres"] = data["genres"].apply(convert)
data["keywords"] = data["keywords"].apply(convert)
data.head()
```

```python
def convert2(obj):
    lst = []
    counter = 0
    for i in ast.literal_eval(obj):     # literal_eval() used to convert string enclosed list to the list
        if(counter != 3):
            lst.append(i["name"])
        else:
            break;
            
    return lst
```

```python
data["cast"] = data["cast"].apply(convert2)
data.head()
```

```python
def fetch_director_from_crew(obj):
    lst = []
    for i in ast.literal_eval(obj):  # literal_eval() used to convert string enclosed list to the list
        if(i["job"] == "Director"):
            lst.append(i["name"])
            break;
    return lst
```

```python
data["crew"] = data["crew"].apply(fetch_director_from_crew)
data.head()
```

```python
data["genres"] = data["genres"].apply(lambda x: [i.replace(" ","") for i in x ])
data["keywords"] = data["keywords"].apply(lambda x: [i.replace(" ","") for i in x ])
data["cast"] = data["cast"].apply(lambda x: [i.replace(" ","") for i in x ])
data["crew"] = data["crew"].apply(lambda x: [i.replace(" ","") for i in x ])
data.head()
```

```python
data["overview"] = data["overview"].apply(lambda x: x.split())
data.head()
```

```python
data["tags"] =  data["overview"] + data["genres"] + data["keywords"] + data["cast"] + data["crew"]
data.head()
```

```python
data = data[["id","title","tags"]]
data.head()
```

```python
data["tags"] = data["tags"].apply(lambda x: [i.lower() for i in x])
data["tags"] = data["tags"].apply(lambda x: " ".join(x))
data.head()
```

```python
# converting every movie to a vector 

from sklearn.feature_extraction.text import CountVectorizer  # This class help us to covert each movie tags to a vector
cv = CountVectorizer(max_features = 4000,stop_words = "english")
```

```python
vectors = cv.fit_transform(data["tags"]).toarray()
vectors.shape
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity    # it gives us value betweeen 0 to 1 , it helps to find out how much the two vectors are similar
similarity = cosine_similarity(vectors)
similarity.shape
```

```python
def recommend(movie):
    movie_index = data[data["title"] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x : x[1])[1:6]
    
    for i in movies_list:
        print(data.iloc[i[0]].title)
    
```

```python
recommend("Batman Begins")
```
